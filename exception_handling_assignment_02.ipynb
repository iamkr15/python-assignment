```python
QUES 3  :>

The ArithmeticError class in Python is a base class for various arithmetic-related exceptions. The primary exceptions that derive from ArithmeticError include:

1)ZeroDivisionError
2)OverflowError
```

```python
1)ZeroDivisionError

try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error: Division by zero is not allowed.", e)
```

```python
2) OverflowError
try:
    result = 10 ** 1000
    print(result)
except OverflowError as e:
    print("Error: Result is too large to be represented.", e)
```

```python
QUES 4  :>

The LookupError class in Python serves as a base class for exceptions that occur when a lookup operation fails.
 It is primarily used to group exceptions that arise from failed searches for items in data structures. 
The two most common exceptions derived from LookupError are KeyError and IndexError.

1)KeyError:> A KeyError occurs when trying to access a dictionary with a key that does not exist.

2)IndexError:>  An IndexError occurs when trying to access an index in a list (or other indexable collection) that is out of range.
```

```python
1)KeyError :>

my_dict={"name":"sudh","age":30}
try:
    value = my_dict["address"]
except KeyError as e:
    print(f"keyerror : {e} key is not available")
```

```python
2)IndexError

my_list=[1,2,3,4,5,6,7,8]
try:
    value = my_list[10]
except IndexError as e:
    print(f"indexerror : {e} ")
```
